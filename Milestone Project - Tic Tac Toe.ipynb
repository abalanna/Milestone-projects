```python
# '''TIC TAC TOE game v. 0.2'''

# import os
print('Welcome to the Tic Tac Toe game!\n')
p1 = input('Player 1: Please choose your symbol (X or O): ')

if p1 == 'X':
    p2 = 'O'
    print('Player 2: Your symbol is: O')
else:
    p2 = 'X'
    print('Player 2: Your symbol is: X')

print('\nThis is your board.\n')


def draw_board():
    '''Returns sample board with cells numbers'''

    db = {789: "     |     |     \n  7  |  8  |  9  \n     |     |     ",
          456: "\n ----+-----+----\n     |     |     \n  4  |  5  |  6  \n     |     |     ",
          123: "\n ----+-----+----\n     |     |     \n  1  |  2  |  3  \n     |     |     "}

    return print(db[789], db[456], db[123])

draw_board()

# cells assigning
c1 = c2 = c3 = c4 = c5 = c6 = c7 = c8 = c9 = ' '

cells = {'1': c1, '2': c2, '3': c3, '4': c4, '5': c5, '6': c6, '7': c7, '8': c8, '9': c9}


def game_board():
    '''Returns current game board'''

    gb = {789: f"\n     |     |     \n  {cells['7']}  |  {cells['8']}  |  {cells['9']}  \n     |     |     ",
          456: f"\n ----+-----+----\n     |     |     \n  {cells['4']}  |  {cells['5']}  |  {cells['6']}  \n     |     |     ",
          123: f"\n ----+-----+----\n     |     |     \n  {cells['1']}  |  {cells['2']}  |  {cells['3']}  \n     |     |     "}

    return print(gb[789], gb[456], gb[123])


# def clear_screen():
#     os.system('cls' if os.name == 'nt' else 'clear')

def p1_move():
    p1_move.counter += 1
    n1 = input('Player 1: Please choose the number to place your symbol: ')
    for key, value in cells.items():
        if n1 == key:
            cells[key] = p1
    return game_board()

def p2_move():
    n2 = input('Player 2: Please choose the number to place your symbol: ')
    for key, value in cells.items():
        if n2 == key:
            cells[key] = p2
    return game_board()
          
p1_move.counter = 0
play = True

while play:
    p1_move()
    # clear_screen()
          
    win = [[cells['1'], cells['2'], cells['3']], 
           [cells['4'], cells['5'], cells['6']], 
           [cells['7'], cells['8'], cells['9']], 
           [cells['7'], cells['4'], cells['1']], 
           [cells['8'], cells['5'], cells['2']], 
           [cells['9'], cells['6'], cells['3']], 
           [cells['7'], cells['5'], cells['3']], 
           [cells['1'], cells['5'], cells['9']]] 
          
    if p1_move.counter <= 2:
        p2_move()
    elif p1_move.counter >= 3:
        for w in win:
            print(win)
            if w == [p1, p1, p1]:
                print('\nCongratulations! The winner is Player 1.')
                break
            else: 
                p2_move()
                print(win)
                if w == [p2, p2, p2]:
                    print('\nCongratulations! The winner is Player 2.')
                    break
        print('lololoLOOOOO')
        break
```
